```python
import pandas as pd
import math
import sklearn
import sklearn.linear_model as sk
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("/content/drive/MyDrive/heart.csv")
```

Kolumny opisują:


**age**,


**sex**,


**chest pain type** (4 values),


**resting blood pressure**,


**serum cholestoral** in mg/dl,


**fasting blood sugar** > 120 mg/dl,


**resting electrocardiographic results** (values 0,1,2),


**maximum heart rate achieved**,


**exercise induced angina**,


**oldpeak** = ST depression induced by exercise relative to rest,


**the slope of the peak exercise ST segment**,


**number of major vessels (0-3) colored by flourosopy**,


**thal**: 0 = normal; 1 = fixed defect; 2 = reversable defect,


**target**: 0= less chance of heart attack 1= more chance of heart attack.

Przewidywaną wartością będzie **target** (0 -zmniejszone ryzyko ataku serca, 1 - zwiększone ryzyko).

Zdecydowałem się na zachowanie wszystkich danych. Ponieważ występują dane w różnych przedziałach zostały one ustandaryzowane.

```python
features = df.drop(columns=['target','sex'])
target = df['target']
```

```python
scaler = StandardScaler()
features_stan = scaler.fit_transform(features)
```

```python
df_stan = pd.DataFrame(features_stan, columns=features.columns)
df_stan['target'] = target
df_stan['sex'] = df['sex']
```

```python
print(df_stan)
```

```python
df_stan.head()
```

```python
train, test = train_test_split(df_stan, test_size=0.33, stratify=df_stan["target"])
print(train.shape, test.shape)
```

```python
train_y=train["target"]
train_x=train.drop(["target"], axis=1)

test_y=test["target"]
test_x=test.drop(["target"], axis=1)
```

```python
model = sk.LogisticRegression()
model.fit(train_x,train_y)
```

```python
print("Wagi:", model.coef_)
print("Odchylenie:", model.intercept_)
```

```python
rezultaty = list(model.predict(test_x))
odpowiedzi = list(test_y)
testy=[]
for i in range(len(rezultaty)):
  testy.append((odpowiedzi[i],rezultaty[i]))
```

```python
tp = testy.count((1,1))
tn = testy.count((0,0))
fp = testy.count((0,1))
fn = testy.count((1,0))
total=(len(testy))
```

```python
print(f"TP = {tp}, TN = {tn}, FP = {fp}, Fn = {fn}, w sumie {total} przypadków")
```

```python
accuracy=(tp+tn)/total
precision=tp/(tp+fp)
recall=tp/(tp+fn)
fscore=(2*precision*recall)/(precision+recall)
sensitivity=tp/(tp+fn)
specificity=tn/(tn+fp)
balanced_accuracy=(sensitivity+specificity)/2
```

```python
print(f"Dokładność:\t\t{accuracy}\t{'%.02f' % (accuracy*100)}% ")
print(f"Precyzja:\t\t{precision}\t{'%.02f' % (precision*100)}% ")
print()
print(f"Czułość:\t\t{sensitivity}\t{'%.02f' % (sensitivity*100)}% ")
print(f"Swoistość:\t\t{specificity}\t{'%.02f' % (specificity*100)}% ")
print(f"Zbalansowana dokładność:{balanced_accuracy}\t{'%.02f' % (balanced_accuracy*100)}% ")
print(f"F-score:\t\t{fscore}\t{'%.02f' % (fscore*100)}% ")
```

W oryginalnej wersji modelu, podział danych treningowych i testowych wynosił 70/30. Biorąc pod uwagę wartości wskaźników model był w pewnym stopniu skuteczny. Wskaźniki dokładności oraz precyzji (oba ~80%) sugerują, że model się sprawdzał, jednak biorąc pod uwagę, że podobne wartości zostały osiągnięte w poprzednich wariantach, oraz około 20-procentową lukę we wskaźnikach, być może na miejscu byłaby weryfikacja wartości kolumn i ograniczenie danych w celu osiągnięcia większej dokładności oraz precyzji.

Przy niewielkim zwiększeniu proporcji danych treningowych do testowych (67/33), wskaźnik dokładności zwiększył się aż o prawie 7 punktów procentowych. Poniżej wersja z uśrednianiem wyników.

```python
def regresja(df_stan):
  train, test = train_test_split(df_stan, test_size=0.33, stratify=df_stan["target"])
  train_y=train["target"]
  train_x=train.drop(["target"], axis=1)

  test_y=test["target"]
  test_x=test.drop(["target"], axis=1)

  model = sk.LogisticRegression()
  model.fit(train_x,train_y)

  rezultaty = list(model.predict(test_x))
  odpowiedzi = list(test_y)
  testy=[]
  for i in range(len(rezultaty)):
    testy.append((odpowiedzi[i],rezultaty[i]))

  tp = testy.count((1,1))
  tn = testy.count((0,0))
  fp = testy.count((0,1))
  fn = testy.count((1,0))
  total=(len(testy))

  accuracy=(tp+tn)/total
  precision=tp/(tp+fp)
  recall=tp/(tp+fn)
  fscore=(2*precision*recall)/(precision+recall)
  sensitivity=tp/(tp+fn)
  specificity=tn/(tn+fp)
  balanced_accuracy=(sensitivity+specificity)/2
  return [accuracy,precision,sensitivity,specificity,fscore,balanced_accuracy]

n=100
wynik=[0,0,0,0,0,0]
for i in range(n):
  results=regresja(df_stan)
  for i,v in enumerate(results):
    wynik[i]+=v

wynik=[w/n for w in wynik]
print(f'''accuracy: {wynik[0]}
precision: {wynik[1]}
sensitivity: {wynik[2]}
specificity: {wynik[3]}
fscore: {wynik[4]}
balanced_accuracy: {wynik[5]}''')

```
