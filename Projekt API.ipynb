```python
import requests
import time, pprint
import jq
import json
```

```python
def api(query, page, language):
  url = "https://openlibrary.org/search.json"
  wynik = []
  parameters = {"q":query, "page":page, "has_fulltext":"true"}

  for i in range (1,11):
    r=requests.get(url, params=parameters)

    dane=r.json()
    wynik+=dane["docs"]
    print("Page",page,"Number of results:",len(dane["docs"]))

    if len(dane["docs"])==0:
      print ("No more results found. Ending loop.\n")
      break
    if len(dane["docs"])!=100:
      print("Loop incomplete. Continuing to next page.\n")
      continue
    print("Loop complete. Continuing to next page.\n")


    w = jq.compile('.[]|{"Title":.title, "Author(s)":.author_name, "Publication date":.first_publish_year, "Language(s)":.language } ').input_value(wynik).all()
    w1 = jq.compile(f'.[] | select(.["Language(s)"] | index("{language}"))').input_value(w).all()

    # return w1 # w celu sprawdzenia potem czy funkcja poprawnie zwraca listę json

    with open("z3-output.json", "w") as plik:
      json.dump(w1, plik)
```

```python
api("melville", "1", "chi")
```
